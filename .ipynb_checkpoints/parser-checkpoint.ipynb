```python
import pathlib

import pandas as pd
import json

# Load the JSON file
result_dir = pathlib.Path()/"results"
instance_names = [
    "C101-10",
    "C101-25",
    "C101-5",
    "C102-25",
    "C103-15",
    "C103-5",
    "C104-10",
    "C106-15",
    "C202-10",
    "C202-15",
    "C205-10",
    "C206-25",
    "C206-5",
    "C208-15",
    "C208-5",
    "R101-25",
    "R102-10",
    "R102-15",
    "R102-25",
    "R103-10",
    "R103-25",
    "R104-25",
    "R104-5",
    "R105-15",
    "R105-5",
    "R201-10",
    "R202-15",
    "R202-5",
    "R203-10",
    "R203-5",
    "R209-15",
    "RC102-10",
    "RC103-15",
    "RC105-5",
    "RC108-10",
    "RC108-15",
    "RC108-5",
    "RC201-10",
    "RC202-15",
    "RC204-15",
    "RC204-5",
    "RC205-10",
    "RC208-5",
]
for instance_name in instance_names:
    filepath = result_dir/(instance_name+".json")
    
    with open(filepath, 'r') as f:
        data = json.load(f)
    
    # To store the dataframes
    dfs = {}
    
    # These fields are long lists you may want as DataFrames or Series
    time_series_keys = [
        'bestBsvCostHistory',
        'bestPreBsvCostHistory',
        # add more keys if needed (e.g. 'bestTotalCostHistory', etc.)
    ]
    
    # Convert each long history list to a DataFrame or Series
    for key in time_series_keys:
        if key in data:
            # As DataFrame with iteration index
            dfs[key] = pd.DataFrame({
                'iteration': range(len(data[key])),
                key: data[key]
            })
    
    # Example for a single-key DataFrame
    # dfs['bestBsvCostHistory'].head()
    
    # If you have tabular lists of dicts, convert them as well
    # Example: If data['operatorUsageHistory'] is a list of dicts
    for key, value in data.items():
        if isinstance(value, list) and len(value) > 0 and isinstance(value[0], dict):
            dfs[key] = pd.DataFrame(value)
    
    # Print summary of all dataframes created
    for name, df in dfs.items():
        print(f"DataFrame: {name}, shape: {df.shape}")
    
    # Access:
    # dfs['bestBsvCostHistory']
    # dfs['operatorUsageHistory'] (if present)
```
